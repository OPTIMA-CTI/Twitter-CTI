```python
from datetime import datetime, timezone
def first_submission(df_row):

  date_format=r'%Y-%m-%d %H:%M:%S%z'
  source_date = datetime.strptime(df_row['tweet_date'], date_format)
  source_date =  source_date.replace(tzinfo=None)
  vt_date = datetime.strptime(df_row['vt_first_seen'], r'%Y-%m-%d %H:%M:%S')
  av_date = datetime.strptime(df_row['av_first_seen'], r'%Y-%m-%d %H:%M:%S')
  if(source_date< vt_date):
    if(source_date< av_date):
       early_report = "Twitter"
    else:
      early_report = "Alien Vault"
  else:
    if(vt_date< av_date):
       early_report = "Virus Total"
    else:
      early_report = "Alien Vault"
  return early_report
```

```python
#url
from matplotlib import pyplot as plt
plt.rcParams['font.size'] = 16
from upsetplot import plot
plot(example,show_counts=True,subset_size='sum')
```
