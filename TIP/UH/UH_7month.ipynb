```python
from urlhaus import UrlHaus
import time
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

start_time = time.time()

'''
    URLhaus report canbe generated by passing ioc, ioc type, source name, date and date format to UrlHaus object

    In this program we import the year.csv from the TweetFeed data (which contains the ioc details obtained from twitter)
    and then passing the values and obtaining the result as a data frame.
    It canbe then stored as a csv file for the ease of data transfer.

    The process here is executed by the help of ThreadPoolExecutor library from concurrent.futures module.
    By setting the number of threads to 50 or above we can run this code and obtain result in a much faser way.

    On a normal execution it takes an average of 1 sec per ioc to complete.
    With Multi Threading same can be obtained with an average of 0.03 sec

    When using ThreadPoolExecutor the result is atleast 30 times faster.

'''

''' 
    Method report will recieves index value and will generated the URLhaus report in the DataFrane corresponding to that index

    ioc, ioc type, date are generated correspondiing to the ioc and passed to UrlHaus object
    then the generated data is retrived by calling the class atribute of UrlHaus
    and finally write that to DataFrame by using 'loc' method

'''
def report(index):
    ioc = df.loc[index, 'ioc_value']
    status=df.loc[index, 'urlhaus_status']
    ioc_type=df.loc[index,'ioc_type']
    if ioc_type not in ['hash']:
      return
    """if status in ['Malicious','NotFound']:
      return"""
    ioc_type = df.loc[index, 'ioc_type']
    source_date = df.loc[index, 'tweet_date']
    source_date=source_date[:19]
    try:
        uh = UrlHaus(ioc, ioc_type, source_date, source_name='Twitter',date_format='%Y-%m-%d %H:%M:%S')
    except Exception as e:
        print(index, ioc, 'Error\n',e)
        df.loc[index, 'urlhaus_status'] = 'Error'
        return

    print(index,source_date, ioc, uh.urlhaus_report)

    df.loc[index, 'urlhaus_status'] = uh.urlhaus_report

    if uh.status:
        df.loc[index, 'url_status'] = uh.status

    if uh.first_submission_date:
        df.loc[index, 'first_submission'] = uh.first_submission_date

    if uh.first_report:
        df.loc[index, 'first_report'] = uh.first_report
        df.loc[index, 'first_report_latency'] = uh.first_report_latency
        df.loc[index, 'first_report_latency(s)'] = uh.first_report_latency_in_sec

    if uh.reference:
        df.loc[index, 'reference'] = uh.reference
    print(df.loc[index, 'first_submission'])
    print(df.loc[index, 'first_report'])
    print(df.loc[index, 'first_report_latency'])
    print(df.loc[index, 'first_report_latency(s)'])
    return


'''
    ioc data is collected from github and get updated with the locally stored output if any
    Then it is processed by the ThreadPoolExecutor and finally output is stored to the local database
'''

fields_name = ['tweet_date', 'account', 'ioc_type', 'ioc_value', 'type_of_attack', 'tweet_url', 'text',
               'urlhaus_status', 'url_status', 'first_submission', 'first_report', 'first_report_latency', 'first_report_latency(s)', 'reference']

#cloud_data = r'https://raw.githubusercontent.com/0xDanielLopez/TweetFeed/master/year.csv'


# cloud_df = pd.read_csv(cloud_data, names=fields_name,encoding='utf-8', low_memory=False) # To read data from cloud

# cloud_df.sort_values(by=['t_time']).to_csv('year.csv', na_rep='', index=None, errors='ignore') # save the cloud data locally

# df = pd.read_csv('D:\Programs\Project\year.csv',names=fields_name, encoding='utf-8', low_memory=False)
# df = pd.read_csv('year.csv', encoding='utf-8', low_memory=False)

df = pd.read_csv('/content/drive/MyDrive/Twitter_Work_Code/TIP/UH/UH_Output_Non_CVE_17_11_2022.csv', engine='python')

# df.update(saved_data) # to update the cloud data with the saved output

# df = df.sort_values(by=['t_time']).reset_index(drop=True)
print(df)

# Multi Threading process happens with an optimal number of workers(max_workers), the optimal number can be varied with the availabilty of internet connectivity speed and the device hardware capacity

try:
    with ThreadPoolExecutor(max_workers=1) as executor:
        executor.map(report, range(1,len(df)))
        executor.shutdown(wait=True)

except Exception as e:
    print(e)

#print(df)

df.to_csv('/content/drive/MyDrive/Twitter_Work_Code/TIP/UH/UH_Output_Non_CVE_17_11_2022.csv', na_rep='', index=None, errors='ignore') # To write the data to csv

end_time = time.time()

print(f'\nTime : {end_time - start_time : .2f}')
```
